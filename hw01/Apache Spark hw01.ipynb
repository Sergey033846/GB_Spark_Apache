Найти самую длинную последовательность упорядоченных чисел в RDD и вывести её.

```python
from pyspark import SparkContext
```

```python
# функция разделения исходной последовательности на "подпоследовательности"
# compare_func - функция сравнения для выбора порядка упорядоченности (например, по возрастанию или по убыванию)
def split_sequence(sequence, compare_func):
  result = []
  temp = [sequence[0]]
  for x in sequence[1:]:
      if compare_func(temp[-1], x):
          temp.append(x)
      else:
          result.append(temp)
          temp = [x]
  if temp:
      result.append(temp)
  return result
```

```python
data = [1, 2, 3, 2, 1, 7, 8, 9, 10, -1, -2, 0, 3, 4, 5]

# формируем "подпоследовательности", упорядоченные по возрастанию
sequences = split_sequence(data, lambda x1, x2: x1 <= x2)
print(sequences)
```

```python
sc = SparkContext("local", "LongestSequence")
rdd = sc.parallelize(sequences)
```

```python
# выбираем самые длинные последовательности
max_len = rdd.map(len).max()
longest_sequences = rdd.filter(lambda x: len(x) == max_len).collect()
print(longest_sequences)
```

```python
sc.stop()
```
